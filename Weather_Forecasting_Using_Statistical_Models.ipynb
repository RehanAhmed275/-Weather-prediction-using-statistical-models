```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
```

```python
# Load the dataset
weather = pd.read_csv("/content/GlobalWeatherRepository.csv")
```

```python
weather
```

```python
# Convert 'last_updated_epoch' to datetime
weather['last_updated_epoch'] = pd.to_datetime(weather['last_updated_epoch'], unit='s')

# Set 'last_updated_epoch' as the index
weather.set_index('last_updated_epoch', inplace=True)

# Check the temperature column
print(weather['temperature_celsius'].describe())

# Selecting only the temperature column as the target variable
weather = weather[['temperature_celsius']]
```

```python
# Print the first few rows of the dataframe
print(weather)
```

#Temporal Analysis

```python
# Aggregate temperature data on a daily basis
daily_temperature = weather['temperature_celsius'].resample('D').mean()  # You can change 'mean' to 'median', 'sum', etc.

# Visualize aggregated temperature data
plt.figure(figsize=(18, 6))
plt.plot(daily_temperature.index, daily_temperature, label='Daily Temperature (Celsius)', color='blue')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.title('Aggregated Daily Temperature Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Visualize the temperature data over time using a scatter plot
plt.figure(figsize=(18, 6))
plt.scatter(weather.index, weather['temperature_celsius'], color='orange', label='Temperature (Celsius)')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Extracting temperature data
temperature = weather['temperature_celsius']

# Splitting the data into train and test sets
train = temperature.iloc[:int(0.8*len(temperature))]
test = temperature.iloc[int(0.8*len(temperature)):]
```

#ARIMA & SARIMA

```python
# Ensure the index is datetime and monotonic
train.index = pd.to_datetime(train.index)
train = train.sort_index()

# Define and fit the ARIMA model
arima_model = ARIMA(train, order=(5,1,0))  # Using train DataFrame directly
arima_fit = arima_model.fit()

# Make predictions
arima_predictions = arima_fit.forecast(steps=len(test))

# Reindex the predictions with the original index
arima_predictions.index = test.index
```

```python
# Make predictions
arima_predictions = arima_fit.forecast(steps=len(test))
```

```python
# SARIMA Model
# Define and fit the SARIMA model
sarima_model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))  # You can adjust the order and seasonal_order parameters
sarima_fit = sarima_model.fit()

# Make predictions
sarima_predictions = sarima_fit.forecast(steps=len(test))
```

#Model Evalution

```python
# Evaluate the models
arima_mse = mean_squared_error(test, arima_predictions)
sarima_mse = mean_squared_error(test, sarima_predictions)

print("ARIMA Mean Squared Error:", arima_mse)
print("SARIMA Mean Squared Error:", sarima_mse)
```

#Temperature prediction using ARIMA and SARIMA

```python
# Plot the actual vs predicted values
plt.figure(figsize=(18, 6))
plt.plot(test.index, test, label='Actual', color='blue')
plt.plot(test.index, arima_predictions, label='ARIMA Predictions', color='red')
plt.plot(test.index, sarima_predictions, label='SARIMA Predictions', color='green')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Prediction using ARIMA and SARIMA')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the actual vs predicted values
plt.figure(figsize=(18, 6))

# Plot actual temperature
plt.plot(test.index, test, label='Actual', color='blue')

# Plot ARIMA predictions
plt.plot(test.index, arima_predictions, label='ARIMA Predictions', color='red')

# Plot SARIMA predictions
plt.plot(test.index, sarima_predictions, label='SARIMA Predictions', color='green')

# Shading for ARIMA predictions
plt.fill_between(test.index, arima_predictions, test, color='red', alpha=0.1)

# Shading for SARIMA predictions
plt.fill_between(test.index, sarima_predictions, test, color='green', alpha=0.1)

plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Prediction using ARIMA and SARIMA')
plt.legend()
plt.grid(True)
plt.show()
```

#Forecast Accuracy Matrics

```python
# Forecast accuracy metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae_arima = mean_absolute_error(test, arima_predictions)
mse_arima = mean_squared_error(test, arima_predictions)
rmse_arima = np.sqrt(mse_arima)
mae_sarima = mean_absolute_error(test, sarima_predictions)
mse_sarima = mean_squared_error(test, sarima_predictions)
rmse_sarima = np.sqrt(mse_sarima)

print("ARIMA Metrics:")
print("MAE:", mae_arima)
print("MSE:", mse_arima)
print("RMSE:", rmse_arima)

print("\nSARIMA Metrics:")
print("MAE:", mae_sarima)
print("MSE:", mse_sarima)
print("RMSE:", rmse_sarima)
```

```python
# Define metrics and models
models = ['ARIMA', 'SARIMA']
mae = [mae_arima, mae_sarima]
mse = [mse_arima, mse_sarima]
rmse = [rmse_arima, rmse_sarima]

# Plot MAE
plt.figure(figsize=(10, 5))
plt.bar(models, mae, color='skyblue')
plt.xlabel('Model')
plt.ylabel('Mean Absolute Error (MAE)')
plt.title('Forecast Accuracy: Mean Absolute Error (MAE)')
plt.show()

# Plot MSE
plt.figure(figsize=(10, 5))
plt.bar(models, mse, color='lightgreen')
plt.xlabel('Model')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Forecast Accuracy: Mean Squared Error (MSE)')
plt.show()

# Plot RMSE
plt.figure(figsize=(10, 5))
plt.bar(models, rmse, color='lightcoral')
plt.xlabel('Model')
plt.ylabel('Root Mean Squared Error (RMSE)')
plt.title('Forecast Accuracy: Root Mean Squared Error (RMSE)')
plt.show()
```

# Ensemble by averaging predictions of ARIMA and SARIMA


```python
# Ensemble by averaging predictions of ARIMA and SARIMA
ensemble_predictions = (arima_predictions + sarima_predictions) / 2

# Evaluate ensemble model
mae_ensemble = mean_absolute_error(test, ensemble_predictions)
mse_ensemble = mean_squared_error(test, ensemble_predictions)
rmse_ensemble = np.sqrt(mse_ensemble)

print("Ensemble Metrics:")
print("MAE:", mae_ensemble)
print("MSE:", mse_ensemble)
print("RMSE:", rmse_ensemble)
```
